## Installing Python via the "Anaconda Installer"

The easiest way to install Python is by installing `Anaconda`, which is a bundle of Python plus many additional goodies that you need for doing data science (e.g. Spyder and Jupyter). Installation instructions are below:

1. Go to: https://www.anaconda.com/products/individual#Downloads
1. Download the “Graphical Installer” for your operating system (i.e. Windows or MacOS)
1. Open the download file and follow instructions
1. Note that you will be downloading the free “Individual Edition” of Anaconda
1. Feel free to search YouTube for tutorials
   * Search for "how to install Anaconda"
   * Sort results by upload date. Choose a recent video


## How to open Spyder

Most likely, most of your Python work will be done in Spyder. We will learn details on how to use Spyder in a later lab. But for now you may want to know how to open it.

1. Type `Spyder` in your Windows search bar or in your Mac's spotlight search bar
1. Click on Sypder
1. You may want to pin Spyder to your takbar so you can open easier next time
